```python
# coding:utf-8

%matplotlib inline     
# 让jupyter显示图像，如果使用别的软件，通过plt.show显示图像

import numpy as np 
import matplotlib as mlp
import matplotlib.pyplot as plt
import seaborn as sns


# 绘制一个sin曲线
def sinplot(flip = 1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
sns.set_style()
sinplot()
```

## 通过set_context()调整绘图元素

sns.set() 可以重置参数

sns.set_context("paper")   四种预设，按线条越来越粗：paper/notebook/talk/poster

```python
# 调整图像大小
plt.figure(figsize=(10, 8)) 

sns.set_style("darkgrid")
sns.set_context("paper")
plt.subplot(221)
sinplot()

sns.set_style("dark")
sns.set_context("notebook")
plt.subplot(222)
sinplot()

sns.set()
sns.set_context("talk")
plt.subplot(223)
sinplot()

sns.set_style("ticks")
sns.set_context("poster")
plt.subplot(224)
sinplot()

```

# 线性关系可视化

## 四个变量关系   绘制多个facet
如： sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", data=tips)

如： sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", row="sex",data=tips)

# 分类数据的绘制
    
将Seaborn的分类图分为三类，将分类变量每个级别没得每个观察结果显示出来，显示每个观察分布的抽象表示，以及应用统计估计显示的权重趋势和置信空间：
    1: swarmplot() 和 stripplot()
    2: boxplot() 和 violinplot()
    3: barplot() 和 pointplt()



## 分类散点图
显示分类变量级别中某些定量变量的值的一种简单方法使用 stripplot（），它会将分散图概括为其中一个变量是分类的。如图1
sns.stripplot(x="day",y="total_bill",data=tips)

但是图1有一个缺点，就是会有部分点重叠的情况，这导致很难看到数据的完整分布，通常简单的解决方法是加入一些随机的“抖动”调整位置，如图2。
sns.stripplot(x="day",y="total_bill",data=tips, jitter=True)

#### 因此，这种情况下，建议使用swarmplot()，它使用避免重叠点的算法将每个散点分开，这种方法比抖动更好，因为抖动也有可能重叠。如图3
sns.swarmplot(x="day",y="total_bill",data=tips)
同样支持绘制多个参数（增加hue参数类别）  同时，如果想将图像横过来，可通过orient参数，推荐直接将x,y参数对调实现。 如图4
sns.swarmplot(x="total_bill",y="day",hue="time",data=tips)

```python
sns.set(style="whitegrid", color_codes=True)
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")

plt.figure(figsize=(15,8))
plt.subplot(221)
sns.stripplot(x="day",y="total_bill",data=tips)

plt.subplot(222)
sns.stripplot(x="day",y="total_bill",data=tips, jitter=True)

plt.subplot(223)
sns.swarmplot(x="day",y="total_bill",data=tips)

plt.subplot(224)
sns.swarmplot(x="total_bill",y="day",hue="time",data=tips)
```

## 点图

pointplot()函数，只绘制点估计和置信区间。 点图连接相同hue的点。
sns.pointplot(x="sex",y="survived",hue="class",data=titanic)

支持使用不同的标记和线条样式来展示不同的hue类别层次。
sns.pointplot(x="class",y="survived",hux="sex",data=titanic,palette={"male":"g","female":"m"},markers=["^","o"],linestyles=["-","--"])

```python
# sns.pointplot(x="sex",y="survived",hue="class",data=titanic)
sns.pointplot(x="class",y="survived",hue="sex",data=titanic,
              palette={"male":"g","female":"m"},markers=["^","o"],linestyles=["-","--"])
```

#  factorplot()   终极函数

默认情况下，factorplot()产生一个pairplot()
sns.factorplot(x="day",y="total_bill",hue="smoker",data=tips)

kind参数可以选择图的种类 
sns.factorplot(x="day",y="total_bill",hue="smoker",data=tips,kind="bar")

如果要更改图形的大小和形状，可以指定对应的size和aspect参数。
sns.factorplot(x="time",y="total_bill",hue="smoker",col="day",data=tips,kind="box",size=4,aspect=.5)

seaborn.factorplot(x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean>, ci=95, n_boot=1000, units=None, order=None, hue_order=None, row_order=None, col_order=None, kind='point', size=4, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True, sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)

Parameters：

x,y,hue 数据集变量 变量名
date 数据集 数据集名
row,col 更多分类变量进行平铺显示 变量名
col_wrap 每行的最高平铺数 整数
estimator 在每个分类中进行矢量到标量的映射 矢量
ci 置信区间 浮点数或None
n_boot 计算置信区间时使用的引导迭代次数 整数
units 采样单元的标识符，用于执行多级引导和重复测量设计 数据变量或向量数据
order, hue_order 对应排序列表 字符串列表
row_order, col_order 对应排序列表 字符串列表
kind : 可选：point 默认, bar 柱形图, count 频次, box 箱体, violin 提琴, strip 散点，swarm 分散点（具体图形参考文章前部的分类介绍）
size 每个面的高度（英寸） 标量
aspect 纵横比 标量
orient 方向 "v"/"h"
color 颜色 matplotlib颜色
palette 调色板 seaborn颜色色板或字典
legend hue的信息面板 True/False
legend_out 是否扩展图形，并将信息框绘制在中心右边 True/False
share{x,y} 共享轴线 True/False
